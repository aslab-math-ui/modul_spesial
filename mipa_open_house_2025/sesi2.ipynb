### Operasi Perbandingan

Operator perbandingan digunakan untuk membandingkan dua nilai atau data. Hasil dari operasi perbandingan ini adalah nilai boolean, yaitu `True` (benar) atau `False` (salah). Berikut adalah beberapa operator perbandingan yang umum digunakan:

* Sama dengan: `==`

* Tidak sama dengan: `!=`

* Lebih besar dari: `>`

* Lebih kecil dari: `<`

* Lebih besar atau sama dengan: `>=`

* Lebih kecil atau sama dengan: `<=`

```python
# sama dengan

print(5 == 5) 

print(5 == 3)
```

```python
# tidak sama dengan

print(5 != 3)

print(5 != 5)
```

### Tipe Data

Ketika kita akan melakukan berbagai macam operasi tentu kita perlu mengetahui apa yang kita operasikan. Pada contoh sebelumnya sempat disinggung `integer`, `float`, dan `string`. Ketiga hal tersebut adalah tipe data dasar yang sering kita temui di dalam pemrograman.

Secara umum tipe data dapat dibagi menjadi dua kategori besar, yaitu tipe data numerik dan tipe data non-numerik yang meliputi string, karakter, dan boolean. Walaupun begitu masih banyak tipe data lain yang lebih kompleks yang tidak akan kita bahas di sini. Berikut adalah penjelasan singkat mengenai tipe data dasar yang sering kita temui:

### For Loop

Konsep loop adalah bagaimana sebuah program dapat mengulang suatu blok kode tertentu selama kondisi yang ditentukan terpenuhi. Ada dua jenis loop yang umum digunakan, yaitu `for` dan `while`. 

Penggunaan `for` digunakan untuk mengulang perintah dengan menggunakan jumlah iterasi yang ditentukan sebelumnya. 

```python
for i in range(5):
    print("Iterasi ke-",i)
```

Kode di atas adalah melakukan perintah pengulangan sebanyak 5 kali dan mencetak perintah ke-$i$ ke layar. Perhatikan bahwa indeks perulangan dimulai dari 0 hingga 4 hal ini karena indeks pada Python dimulai dari 0, sehingga total iterasi adalah 5 kali dimulai dari 0. Catatan, ada bahasa pemrograman lain indeks perulangan mungkin dimulai dari 1.

Bagaimana jika kita ingin mencetak bilangan dari 1 sampai 5? Kita bisa melakukannya sebagai berikut:

```python
for i in range(5):
    print("Iterasi ke-", i+1)
```

Jadi di sini kita tambahkan saja satu agar output mencetak dari 1 sampai 5. Misalkan kita ingin menjumlahkan bilangan dari 1 sampai 5, kita bisa melakukannya seperti ini:

```python
hasil = 0
for i in range(5):
    hasil = hasil + (i + 1)
    print("Iterasi ke-", i+1, "hasilnya: ", hasil)

# atau tambahkan saja iterasinya sampai 6 agar kita tidak perlu menambahkan i dengan 1
print("==========================")
hasil_new = 0
for i in range(6):
    hasil_new = hasil_new + i
    print("Iterasi ke-", i, "hasilnya: ", hasil_new)
```

```python
# contoh penggunaan loop untuk menjumlahkan bilangan genap dari 1 sampai 10

hasil_genap = 0
for i in range(11):
    if i % 2 == 0:
        hasil_genap = hasil_genap + i
        print("Menambahkan", i, "hasil sementara:", hasil_genap)

# atau dengan stepping di range
print("==========================")
hasil_genap_step = 0
for i in range(0, 11, 2):
    hasil_genap_step = hasil_genap_step + i
    print("Menambahkan", i, "hasil sementara:", hasil_genap_step)
```

Perhatikan bahwa kita bisa melakukan dua macam bentuk untuk menjumlahkan bilangan genap dari 1 hingga 10. Loop pertama adalah bagaimana kita memberkan syarat di dalam loop ketika suatu bilangan merrupakan bilangan genap. 

Pada loop kedua kita mengatur kondisi `for` untuk melakukan *step* atau loncatan sebesar 2, sehingga hanya bilangan genap saja yang akan diiterasi.

```python
hasil_genap = 0
for i in range(11):
    if i % 2 == 0:
        print("Bilangan genap:", i)

# atau dengan stepping di range
print("==========================")
hasil_genap_step = 0
for i in range(0, 11, 2):
    print("Bilangan genap:", i)
```

### While Loop

Berbeda dengan `for` yang mengulang berdasarkan jumlah iterasi tertentu, `while` mengulang blok kode selama kondisi logika tertentu bernilai `True`. Sederhananya, selama kondisi yang diberikan pada `while` terpenuhi, maka blok kode di dalamnya akan terus dijalankan.

```python
ptr = 1

while (ptr <= 5):
    print("Nilai :", ptr)
    ptr = ptr + 1
```

```python
# jika kita mulai dari angka -5

ptr = -5

while (ptr <= 5):
    print("Nilai :", ptr)
    ptr += 1
```

Oke, jelas sekali perbedaan `while` dari `for` sebelumnya. Apa yang terjadi jika kita tidak memberikan kondisi `ptr` di awal?

```python
while ptr_new <= 5:
    print("Nilai ptr_new:", ptr_new)
    ptr_new = ptr_new + 1
```

Yup jika kita tidak menginisialisasi syarat awal loop berjalan maka `while` akan menghasilkan error dalam hal ini, karena variabel `ptr_new` belum didefinisikan sebelumnya.

```python
angka = 0

while angka != 10:
    print("Angka sekarang:", angka)
    angka = angka + 2

print("======================")

angka_lain = 10

while angka_lain != 0:
    print("Angka sekarang:", angka_lain)
    angka_lain= angka_lain - 2
```

## Fungsi dan Rekursi

Salah satu aspek yang juga sangat penting untuk digunakan dalam pemrograman adalah Fungsi. Fungsi di pemrograman bisa dikatakan memiliki kemiripan dengan fungsi yang ada di matematika. Kita dapat memberikan aturan yang menerima satu atau beberapa nilai sebagai input lalu menghasilkan output. Fungsi adalah blok instruksi yang diberi nama dan bisa digunakan berkali-kali.

Misalkan kita ingin membuat fungsi yang menerima input bilangan real dan mengeluarkan output kuadrat dari bilangan input atau kurang lebih adalah fungsi $f(x) = x^2$ di Python.

```python
def func(x):

    y = x**2

    return y
```

```python
func(2)
```

Kita juga bisa memberikan sebuah aturan yang menerima panjang dan lebar dari persegi panjang dan memngeluarkan luas dari persegi panjang tersebut.

```python
def cari_luas(panjang, lebar):

    luas = panjang * lebar

    return luas
```

```python
cari_luas(10, 8)
```

Misalkan kita ingin menjumlahkan $n$ bilangan pertama.

```python
def jumlahkan(n):
    
    hasil = 0

    for i in range(1, n+1):

        hasil = hasil + i

    return hasil
```

```python
jumlahkan(100)
```

**Rekursi** adalah proses pengulangan sesuatu dengan cara memanggil dirinya sendiri dari versi sebelumnya. Pikirkan ini menjadi sebuah refleksi diri di mana diri kita saat ini adalah produk dari diri kita di masa lalu (kelazzz). **Fungsi Rekursif** adalah fungsi yang memanggil dirinya sendiri untuk mengeluarkan output.

Contoh paling mudah yang bisa kita pikirkan adalah bilangan fibonacci. Bilangan fibonacci ke-$n$ adalah jumlahan dari dua bilangan sebelumnya.

$$
fibonacci\\

0, 1, 1, 2, 3, 5, 8, 13, 21, \ldots
$$

atau secara matematis bisa kita tulis sebagai

$$
\begin{gather}
    F_0 = 0,\, F_1 = 1\\
    F_n = F_{n-1} + F_{n-2}
\end{gather}
$$

```python
def fibonacci(n):

    if (n == 0):
        return 0
    
    elif (n == 1):
        return 1
    
    else:
        return fibonacci(n-1) + fibonacci(n-2)
```

```python
fibonacci(10)
```

Selalu hati-hati dalam menggunakan fungsi rekursif karena ketika kita tidak mendefinisikan semua kasus yang mungkin terjadi (bisa disebut *base case*) fungsi akan terus berjalan hingga kompilasi program mengalami error atau kasus terburuknya perangkat kita mengalami *crash*.
